# ***Q1 Outlier Detection for Wages***


***How can we identify outliers in the Wage column using statistical methods?
Describe the steps, process, and logic behind the chosen method (e.g., IQR or Z-score).
What insights can be derived from these outliers regarding player valuation?***

*we used the IQR method to detect outliers, in the wages column*

```python
def outlier_detector(df,col):
    Q1=df[col].quantile(0.25)
    Q3=df[col].quantile(0.75)
    IQR=Q3-Q1
    lb=Q1-1.5*IQR
    ub=Q3+1.5*IQR
    return df[(df[col]<lb)|(df[col]>ub)]
    
outlier_detector(df,"Wage")
```

```python
n=outlier_detector(df,"Wage")
n["Wage"]
```

```python
plt.figure(figsize=(10,4))
sns.boxplot(x=n["Wage"],color="g")
plt.title("Box plot of Wages with outliers")
plt.xlabel("Wage")
plt.ylabel("Players")
plt.grid(True)
plt.show()
```

**In conclusion , from this visualising boxplot we can see the most of the players have low wages, and only a few players earn much higher wages these are the outliers**

**we calculate skewness to see if the distribution is symmetric, and kurtosis to understand how peaked or falt the distribution is..**

**step-3 (Fit a theoretical distribution)**

```python
mu, std = norm.fit(df["Potential"])

plt.figure(figsize=(8, 6),dpi=120)
sns.histplot(df["Potential"],kde=True, stat="density", label="Actual Distribution")
xmin, xmax = plt.xlim()
x = np.linspace(xmin, xmax, 100)
p = norm.pdf(x, mu, std)
plt.plot(x, p, 'r', linewidth=2, label="Normal Fit")
plt.title("Distribution of Potential with Normal Fit")
plt.xlabel("Potential")
plt.ylabel("Density")
plt.legend()
plt.show()
```

***This histogram visualizes the distribution of players' potential scores in the dataset. The blue bars represent the actual distribution of the data, while the red curve shows a fitted normal distribution. By comparing both, we can observe whether the player potential values approximately follow a bell-shaped (normal) distribution pattern.***

## ***Q3. Normal vs. Student’s t-Distribution***

***Compare the Normal and Student’s t-distributions using the Potential column. Explain the theoretical differences between these distributions and illustrate how they apply to the Potential data. What implications do these differences have for small sample size***

```python
x = np.linspace(df["Potential"].min(), df["Potential"].max(),100)
plt.plot(x, p, 'r', linewidth=2, label="Normal Fit")
plt.plot(x, stats.t.pdf(x, df=len(df["Potential"])-1, loc=df["Potential"].mean(), scale=df["Potential"].std()), label="Student's t")
plt.legend()    
plt.show()
```

***This analysis compares the Normal distribution and Student’s t-distribution using the Potential data. Both curves were plotted to observe their differences. While the normal distribution assumes a fixed bell shape, the student t-distribution has heavier tails and is more suitable for smaller sample sizes.***

# ***Q4. Normal VS. Standard Normal Distribution***

***Explain the difference between a normal distribution and a standard normal distribution using the Potential column. Standardize the Potential data and compare it to the standard normal distribution. How does this transformation aid in statistical analysis?***

**standardize Potential**

```python
df["Potential"]
```

```python
standardized = (df["Potential"] - df["Potential"].mean()) / df["Potential"].std()

x = np.linspace(-3, 3,100)
plt.plot(x, stats.norm.pdf(x),color="red", label="Standard Normal")
sns.histplot(standardized, kde=True, stat="density")
plt.legend()
plt.show()
```

***Here we standardized the Potential values. This means the data was transformed so that it has a mean of 0 and a standard deviation of 1. The red curve represents the Standard Normal distribution, while the bars show the standardized Potential data. From the plot, we can observe that the standardized data roughly follows the Standard Normal distribution.***

# ***Q5 Confidence Intervals for Potential, Wage, and Weight***


*Calculate the 90%, 95%, and 99% confidence intervals for the mean of the Potential, Wage, and Weight columns. What do these intervals indicate about the population parameters, and how does the confidence level affect the interval width?*

```python
df[["Potential", "Wage", "Weight"]].dtypes
```

*converting weight column from string (lbs) to numeric for statistical analysis*

```python
df["Weight"] = df["Weight"].str.replace("lbs","")
df["Weight"] = df["Weight"].astype(float)
```

```python
df["Weight"]
```

```python
columns = ["Potential", "Wage", "Weight"]

for col in columns:
    data = df[col].dropna()
    
    ci90 = stats.t.interval(0.90, len(data)-1, loc=data.mean(), scale=stats.sem(data))
    ci95 = stats.t.interval(0.95, len(data)-1, loc=data.mean(), scale=stats.sem(data))
    ci99 = stats.t.interval(0.99, len(data)-1, loc=data.mean(), scale=stats.sem(data))

    print(f"\n{col}")
    print("90% CI:", ci90)
    print("95% CI:", ci95)
    print("99% CI:", ci99)
```

***We calculated the 90%, 95%, and 99% confidence intervals for the mean of the Potential, Wage, and Weight columns. These intervals show the range where the true average value of the population is likely to lie. As the confidence level increases, the interval becomes wider.***

# ***Q6 Validation of the Central Limit Theorem*** 

*Demonstrate the Central Limit Theorem using the Potential column. How does the distribution of sample means behave as sample size increases? Provide visualizations to support the analysis and explain the implications of the CLT for statistical inference.*

```python
sample_means = [df["Potential"] .sample(n=30, replace=True).mean() for _ in range(1000)]
plt.hist(sample_means, bins=30, density=True,color="skyblue")
plt.title("Distribution of Sample Means",fontsize=12,fontweight="bold")
plt.xlabel("Sample Mean",fontsize=12,fontweight="bold")
plt.ylabel("Frequency",fontsize=12,fontweight="bold")
plt.show()
```

***Here we took many random samples of 30 players and calculated their average Potential. The histogram shows the distribution of these sample means. We can see the shape looks approximately like a normal curve, which demonstrates the Central Limit Theorem.***
